## **Demo/Usage of `excel_sum`**
### 1. With numeric arguments alone



```python
from excel_math_function import excel_sum
excel_sum(7,9,6,4)
```

### 2. With numeric arguments and string-like numbers


```python
excel_sum(22,"45","7.3",12)
```

### 3. With wrong datatype as argument


```python
excel_sum(22,"45","7.3",12,"rt")
```

### 4. With range of cells as argument


```python
# Creating a dataframe as a representation of range of cells
import pandas as pd 
cell_range = pd.DataFrame(
    {
    'A': ['4', '78', 'ade'], 
     'B': ['43', 'gg', '4'], 
     'C': ['8', 'e', '9']
     },
     index=[1,2,3]
)
# showing the dataframe 
display(cell_range)
# using excel_sum function on the range 
excel_sum(cell_range)
```
